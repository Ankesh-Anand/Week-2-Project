# Question 1 scraping Data Analyst Job title, company name, experience and location

```python
driver.get("https://www.naukri.com/")

#putting designation as Data Analyst
designation = driver.find_element(By.CLASS_NAME, "suggestor-input")
designation.send_keys('Data Analyst')

#setting location as Bangalore
location =driver.find_element(By.XPATH, "/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys ("Bangalore")
location1 =driver.find_element(By.XPATH, "/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location1.send_keys ("/Bengaluru")

search = driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()

job_titles=[]
job_locations =[]
```

```python
title=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title[0:10]:
    job_titles.append(i.text)
job_titles
```

```python
firms =[]
experience_required = []

location=driver.find_elements(By.XPATH, '//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in location [0:10]:
    job_locations.append(i.text)

name=driver.find_elements(By.XPATH, '//a[@class="subTitle ellipsis fleft"]')
for i in name [0:10]:
    firms.append(i.text)


exp=driver.find_elements(By.XPATH, '//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in exp [0:10]:
    experience_required.append(i.text)

job_locations
firms
experience_required
```

```python
df=pd.DataFrame({'Job Title':job_titles, 'Job Location': job_locations, 'Company Name':firms ,'Experience Required': experience_required  })
df
```

```python
brandname=[]
prdtdes=[]
price=[]
discount=[]

start=0
stop=3

for page in range(start,stop):
    
    brand=driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    product=driver.find_elements(By.XPATH, '//a[@class="IRpwTa"]')
    price1=driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    disc=driver.find_elements(By.XPATH, '//div[@class="_3Ay6Sb"]')
    
    for b in brand:
        brandname.append(b.text)
    for p in product:
        prdtdes.append(p.text)
    for r in price1:
        price.append(r.text)
    for d in disc:
        discount.append(d.text)

    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)

# print(len(brandname), len(prdtdes), len(price), len(discount))

df=pd.DataFrame({'Brand Name': brandname,'Product Description':prdtdes,'Price': price,'Discount':discount})
df[:100]





```

# Scrape 100 reviews data from flipkart.com for iphone11 phone

```python
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market')
```

```python
print("The description of Iphone 11 reviews")


ratings=[]
reviewsummary=[]
fullreview=[]

# start=0
# end=9

for page in range(10):
    rating=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    review=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    full=driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for r in rating:
        ratings.append(r.text)
    for e in review:
        reviewsummary.append(e.text)
    for f in full:
        fullreview.append(f.text)
# print(len(ratings),len(reviewsummary),len(fullreview))



df=pd.DataFrame({'Ratings':ratings,'Review Summary':reviewsummary,'Full Review':fullreview})
df
```
