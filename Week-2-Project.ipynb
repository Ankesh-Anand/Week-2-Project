# Question 1

Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
driver.get("https://www.naukri.com/")
```

```python
designation = driver.find_element(By.CLASS_NAME, "suggestor-input")
designation.send_keys('Data Analyst')

loc =driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
loc.send_keys ("Bangalore")

search = driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()

job_title =[]
job_location =[]
company_name =[]
experience_required = []

title = driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title[0:10]:
    job_title.append(i.text)
location = driver.find_elements(By.XPATH, '//li[@class="fleft grey-text br2 placeHolderLi location"]')
for j in location [0:10]:
    job_location.append(j.text)
name = driver.find_elements(By.XPATH, '//a[@class="subTitle ellipsis fleft"]')
for k in name [0:10]:
    company_name.append(k.text)
experience = driver.find_elements(By.XPATH, '//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for l in experience [0:10]:
    experience_required.append(l.text)
    
# print(len(job_title), len(job_location), len(company_name), len(experience_required)

df=pd.DataFrame({'Job Title':job_title,'Job Location':job_location,'Company Name':company_name,'Experience Required':experience_required})
df[:10]
```

```python
driver.get('https://www.flipkart.com/')
```

```python
brand_name=[]
product_description=[]
price=[]
discount=[]

# start=0
# stop=3

for page in range(4):
    
    brand=driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    for b in brand:
        brand_name.append(b.text)

    product=driver.find_elements(By.XPATH, '//a[@class="IRpwTa"]')
    for p in product:
        product_description.append(p.text)

    value=driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    for v in value:
        price.append(v.text)
    disc=driver.find_elements(By.XPATH, '//div[@class="_3Ay6Sb"]')
    for d in disc:
        discount.append(d.text)

#     next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
#     next_button.click()
#     time.sleep(3)

print(len(brand_name), len(product_description), len(price), len(discount))

df=pd.DataFrame({'Brand Name': brand_name,'Product Description':product_description,'Price': price,'Discount':discount})
df[:100]
```

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpage https://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
# opening HTML link on automated test software window using driver.get
driver.get("https://www.azquotes.com/")

quotes=driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
quotes.click()
```

```python
quote=[]
author=[]
category=[]

for page in range(10):
    quote1=driver.find_elements(By.XPATH,'//a[@class="title"]')
    author1=driver.find_elements(By.XPATH,'//div[@class="author"]')
    category1=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for q in quote1:
        quote.append(q.text)
    for a in author1:
        author.append(a.text)
    for c in category1:
        category.append(c.text)
        
# print(len(quote),len(author),len(category))

print("Top 1000 Quotes Of All Time")
df=pd.DataFrame({"Quote":quote, "Author":author, "Type of Quotes":category})
df
```
